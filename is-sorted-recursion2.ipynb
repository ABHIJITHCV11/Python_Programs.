```python
arr1 = [1,12,3,4,5,8,9]
arr2 = [1,2,3,4,5,6,7,8]
arr3 = [1,2,3,5,5,6,7,8] # Even tho the array is sorted it wont work
# fix arr3 in git and refactor
```

```python
def is_sorted(a,k=0): # k --> the length of the array
    if k==len(a)-1: # when we reach the last index the array is sorted
        return True    
    return a[k] < a[k+1] and is_sorted(a,k+1) # did not use string slicing
```

```python
is_sorted(arr1)
```

```python
is_sorted(arr2)
```

```python
is_sorted(arr3)
```
