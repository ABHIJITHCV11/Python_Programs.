```python
def recursive_zero_to_nth_element(n):
    if n==0:
        print(n)
        return None
    recursive_zero_to_nth_element(n-1)
    print(n) # Prints while backtracking as the function call is still goin on,function resting in the stack
    # with the same value of n.
```

```python
recursive_zero_to_nth_element(2)
```
