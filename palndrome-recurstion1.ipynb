```python
a = 'malayalam'
b = 'asd'
c = 'add'
d = 'cac'
e = 'lol'

s=[a,b,c,d,e]
```

```python
def pal(s):
    if len(s) == 1:
        print("yes")
        return True
    elif s[0] != s[-1]:
        print("no")
        return False
    else:
        return pal(s[1:-1])
```

```python
for i in s:
    print(i)
    pal(i)
```
